```python
npz_path = "/home/chy/archive-data/processed/unity-class-shuffled-npz"
ugen = UnityEyePoseGenerator(npz_path, batch_size=512, is_item_mode=True)
```

```python
def unity_item_to_iris_item(item: UnityItem):
    frame = item.get_decoded_frame()
    ic = IrisContourItem()
    ic.frame_encoded = item.image_encoded
    ic.frame_origin_shape = frame.shape
    contours = item.iris[:,:2].T
    ic.calculate_ellipse(contours)
    
    return ic
```

```python
def convert_unity_to_iris_npz(gen: UnityEyePoseGenerator, npz_dir_path, npz_size=512):
    block_count, block = 3106, []
    for i in range(3106, len(gen)):
        items = gen[i]
        for item in items:
            try:
                ic_item = unity_item_to_iris_item(item)
                block.append(ic_item)
            except Exception as e:
                continue

            if len(block) == npz_size:
                IrisContourItem.to_npz(npz_dir_path, block, block_count)
                print("*** archived npz[{}] ***".format(block_count))
                block, block_count = [], block_count + 1

    IrisContourItem.to_npz(ic_item_npz_path, block, block_count)
```

### b. vcone (crowdworks) to iris contour item

```python
def vc_js_to_eye_points_and_centers(js_path):
    js = load_json(js_path)
    contours = js['st_eye_contour']
    contours = [(c['x'], c['y']) for c in contours]
    contours = np.array(contours)
    er_ps, el_ps = contours[:19], contours[19:]
    er_ce, el_ce = js['st_eye_center']
    er_ce, el_ce = (er_ce['x'], er_ce['y']), (el_ce['x'], el_ce['y'])
    
    return (er_ps, er_ce), (el_ps, el_ce)

def vc_to_iris_item(js_path, img_path):
    frame_encoded = read_img_to_byte_arr(img_path)
    frame = decode_img(frame_encoded)
    
    assert frame is not None, "frame should not none"
    
    icr = IrisContourItem()
    icr.frame_encoded = frame_encoded
    icr.frame_origin_shape = frame.shape
    icl = IrisContourItem()
    icl.frame_encoded = frame_encoded
    icl.frame_origin_shape = frame.shape
    
    (er_ps, er_ce), (el_ps, el_ce) = vc_js_to_eye_points_and_centers(js_path)
    
    icr.calculate_circle(er_ps, er_ce)
    icl.calculate_circle(el_ps, el_ce)
        
    return icr, icl
```

```python
def convert_vc_to_iris_npz(jips, npz_dir_path, npz_size=512):
    block_count, block = 0, []
    for jip in jips:
        try:
            icl, icr = vc_to_iris_item(*jip)
        except Exception as e:
            print("Exception[{}] : {} ".format(block_count, e))
            continue
            
        block.append(icl)
        block.append(icr)

        if len(block) == npz_size:
            IrisContourItem.to_npz(npz_dir_path, block, block_count)
            print("*** archived npz[{}] ***".format(block_count))
            block, block_count = [], block_count + 1
```
