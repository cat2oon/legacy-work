# resources

```python
npz_path = "/home/chy/archive-data/processed/everyone-npz-profile"
candide_path="/home/chy/archive-model/candide/candide.npz"
land_mark_path="/home/chy/archive-model/candide/shape_predictor_68_face_landmarks.dat"
model_path = "/home/chy/archive-model/incubator/ove/exp-home2-model.json"
weight_path = "/home/chy/archive-model/incubator/ove/exp-home2-24e-weights.hdf5"
save_dir_path = "/home/chy/archive-data/processed/every-eye"
```

# components

```python
candide = Candide(candide_path, land_mark_path)

intrinsic_param = IntrinsicParameters(1221, 1222, 598, 804)
screen = Screen(1200, 1600)
camera = Camera(4.7, intrinsic_param, screen)

opt_predictor = OpticalAxisPredictor()
opt_predictor.load_model(model_path, weight_path)
face_model = FaceModel(candide)
face = Face(camera, face_model, opt_predictor)
```

# crop eye each profile

```python
profile_json_path = ("{}/profile-list.json".format(npz_path))
profile_js = open(profile_json_path).read()
profile_ids = json.loads(profile_js)
```

```python
for pid in profile_ids:
    counter = 0
    profile_path = "{}/{}/".format(save_dir_path, pid)
    mkdir(profile_path)
    gen = EveryoneFaceProfileGenerator(npz_path, face, profile_id=int(pid))
    
    for l_eye, r_eye in gen:
        if l_eye is None or r_eye is None:
            continue
        l_path = "{}/l-{}.jpg".format(profile_path, counter)
        r_path = "{}/r-{}.jpg".format(profile_path, counter)
        counter += 1
        cv2.imwrite(l_path, l_eye)
        cv2.imwrite(r_path, r_eye)
```
