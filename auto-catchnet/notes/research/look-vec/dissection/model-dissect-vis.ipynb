```python
import sys
from quiver_engine import server

from keras.models import load_model
from keras.engine.saving import model_from_json
```

# load model

```python
def load_model(model_path, weight_path):
    with open(model_path, "r") as json_file:
        model_json = json_file.read()
        net = model_from_json(model_json)
    net.load_weights(weight_path)
    return net
```

```python
model_path = "/home/chy/dev-bench/auto-catch/notes/research/look-vec/lvm-pilot-hard-aug.json"
weight_path = "/home/chy/archive-model/incubator/lookvec/lv-04-0.0070833.hdf5"
m = load_model(model_path, weight_path)
```
